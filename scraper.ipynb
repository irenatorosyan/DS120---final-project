```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup

from pprint import pprint
import time

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

from re import search

from recipe_scrapers import scrape_me

import re
```

```python
titles_ = []
ingredients_ = []
links_ = []

# scraping titles, ingredients and urls from subcategories of subcategories of all recipes
# for example Breakfast -> HEALTHY BREAKFAST RECIPES
def low_scrape(low_url):

    links = []

    url = low_url
    response = requests.get(url)
    page = BeautifulSoup(response.content, "html.parser")


    raw_links = page.find_all('a', class_ = "comp card")

    for i in raw_links:
        if search("https://www.simplyrecipes.com/", i.get('href')):
            if (i.get('href') != 'https://www.simplyrecipes.com/'):
                links.append(i.get('href'))
                
    for j in links:
        scraper = scrape_me(j)
        titles_.append(scraper.title())
        try:
            ingredients_.append(scraper.ingredients())
        except:
            ingredients_.append("None")
        links_.append(j)
```

```python
# scraping titles, ingredients and urls from subcategories of all recipes
# for example Breakfast 
def high_scrape(high_url):
    links1 = []

    url = high_url
    response = requests.get(url)
    page = BeautifulSoup(response.content, "html.parser")


    raw_links = page.find_all('a', class_ = "text-link section__title")

    for i in raw_links:
        links1.append(i.get('href'))

    for a in links1:
        low_scrape(a)
```

```python
# saving the data as a csv file

allrecipes.drop_duplicates().to_csv('allrecipes.csv', index=False)
```
